```python
# Functions to create boolean vectors for segmenting 
# the tweet data based on tweet content:

def photo_embedded(media_list):
    try:
        for m in media_list:
            if m['type'] == 'photo':
                return True
    except TypeError: 
        return False
    
def get_boolvecs_phu(tweets):
    """
    Segment the data into disjoint sets depending on their features.

    'tweets' is a Pandas DataFrame of tweets loaded from a json file.
    """

    contains_photo = [photo_embedded(ml) for ml in tweets['media']]
    contains_hashtags = [h != [] for h in tweets['hashtags']]
    contains_url = [any(u) for u in tweets['urls']]
    all_boolvec = zip(contains_photo, contains_hashtags, contains_url)

    contains_photo_only    = [p and all((not h,not u)) for p,h,u in all_boolvec]
    contains_hashtags_only = [h and all((not p,not u)) for p,h,u in all_boolvec]
    contains_url_only      = [u and all((not p,not h)) for p,h,u in all_boolvec]

    contains_hashtags_url_only   = [all((h,u)) and not p for p,h,u in all_boolvec]
    contains_photo_url_only      = [all((p,u)) and not h for p,h,u in all_boolvec]
    contains_photo_hashtags_only = [all((p,h)) and not u for p,h,u in all_boolvec]

    contains_all =  [all((p,h,u)) for p,h,u in all_boolvec]
    contains_none = [any(phu) == False for phu in zip(contains_photo, contains_hashtags, contains_url)]
    
    return contains_photo_only, contains_hashtags_only, contains_url_only, \
        contains_hashtags_url_only, contains_photo_url_only, contains_photo_hashtags_only, \
        contains_all, contains_none
```

```python
# What content characterizes tweets with:
# (A) 0 retweets (B) between 1 and 5 retweets (C) >5 retweets?

# note: 'pie' plots in pandas.Series.plot() require pandas v0.14 or later.

# get boolean vectors to segment the tweets based on retweet count:
retweets_0 = tweets_sq['n_retweets'] == 0
retweets_1to5 = [all((t>0, t<=5)) for t in tweets_sq['n_retweets']]
retweets_morethan5 = tweets_sq['n_retweets'] > 5

# i.   segment the tweet set based on retweet count
# ii.  takes subsets of each of these based on the tweet content
# iii. count the total in each subset:
typecounts_0retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_0])]
typecounts_1to5retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_1to5])]
typecounts_morethan5retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_morethan5])]

# make the pie charts:
idx = ['photo only', 'hashtags only', 'url only', 'hashtags & url', \
       'photo & url', 'photo & hashtags', 'all 3', 'none']
clr = ['b','y','r','orange','m','g','k','w']

plt.figure(figsize=(4,4))
plt.pie(typecounts_0retweets, labels=idx, colors=clr)
plt.title('0 retweets')

plt.figure(figsize=(4,4))
plt.pie(typecounts_1to5retweets, labels=idx, colors=clr)
plt.title('1 to 5 retweets')

plt.figure(figsize=(4,4))
plt.pie(typecounts_morethan5retweets, labels=idx, colors=clr)
plt.title('>5 retweets')
```

```python
# note: 65 of the 315 tweets_ids here have no match due to a 
# possible Pandas bug (see above).

import numpy

# get a DataFrame of all tweets with >5 rewteets:
# set tweet ID as the indix in the dataframe:
tweets_sq_morethan5 = tweets_sq[retweets_morethan5]
tweets_sq_morethan5 = tweets_sq_morethan5.set_index('tweet_id')
tweets_sq_morethan5['summedDC'] = numpy.nan

# load the data you retreived from the Twitter API:
# (contains info about the number of followers each retweeter has)
dlist = load_json('retweeter_info_total.json')
tweetID_summedDC = {d['tweet_id']:sum(d['retweeters'].values()) for d in dlist}
```

```python
# plot a histogram for our "summed followers" calculation:

fig, axes = plt.subplots(ncols=2)
fig.set_size_inches(15,4)

summedDC_df = pandas.DataFrame.from_dict(tweetID_summedDC, orient='index')
summedDC_df.columns = ['summedDC']
more_than_100k = summedDC_df['summedDC'] > 100000 

summedDC_df['summedDC'].hist(bins=50, ax=axes[0])
axes[0].set_title('tweets with retweets > 5')
axes[0].set_xlabel('summed followers')
axes[0].set_ylabel('count')

summedDC_df['summedDC'][more_than_100k].hist(bins=50, ax=axes[1])
axes[1].set_title('tweets with retweets > 5 and "summed followers" > 100k')
axes[1].set_xlabel('summed followers')
axes[1].set_ylabel('count')
```

```python
# add the data from the file to our data frame:
nomatch_count = 0
for tid in tweets_sq_morethan5.index:
    try:
        tweets_sq_morethan5['summedDC'][tid] = tweetID_summedDC[str(tid)]        
    except KeyError:
#         print 'No tweet_id match for %i (possible Pandas bug).' % tid
        nomatch_count += 1

print 'No tweet_id match for %i tweets (possible Pandas bug).' % nomatch_count

# drop the row where no tweet_id match was found:
tweets_sq_morethan5 = tweets_sq_morethan5.dropna(subset=['summedDC'])
```

```python
# Of the remaining 250 tweets, 156 are actually retweets from other accounts.
# Though these have >0 retweets accoring to twitter_posts.json, no 
# retweet information is returned when quering the Twitter API with their
# tweet_id number.
# Getting information about the retweeters will therefore be slightly 
# more complicated. Drop these for now.

# take the subset of posts that were originally created and tweeted by 
# TheSquareFilm (i.e., are not retweets from other users' tweets):
tweets_sq_morethan5_orig = tweets_sq_morethan5[tweets_sq_morethan5['summedDC'] != 0]
```

```python
# note: in twitter_posts.json, there are tweets that have tweet_id 
# numbers and  n_retweets > 0, but are actually things that TheSquareFilm
# has retweeted. So when you actually send GET requests about these
# followers, you get no information back (& appear to have no retweeters
# from this account).
```

```python
# group the data into 10 disjoint quantiles based on summedDC:

quantile_lowerlimits = [i/10.0 for i in range(10)]
quantilegroups_boolvecs = []

for ql in quantile_lowerlimits:
    quantilelim_lower = tweets_sq_morethan5_orig.quantile(ql)['summedDC']
    quantilelim_upper = tweets_sq_morethan5_orig.quantile(ql+0.1)['summedDC']

    if ql != max(quantile_lowerlimits):
        quantilegroups_boolvecs.append([sdc >= quantilelim_lower and sdc < quantilelim_upper \
                                  for sdc in tweets_sq_morethan5_orig['summedDC']])
    else:
        quantilegroups_boolvecs.append([sdc >= quantilelim_lower and sdc <= quantilelim_upper \
                                  for sdc in tweets_sq_morethan5_orig['summedDC']])
        
# A list of 10 DataFrames containing the data broken into 10 quantiles
# on the summedDC calculation:
quantiles = [tweets_sq_morethan5_orig[bv] for bv in quantilegroups_boolvecs]
```

```python
typecounts = [[sum(s) for s in get_boolvecs_phu(q)]
              for q in quantiles]

# convert from counts to fractions of each category:
typefrac = [[t/float(sum(q)) for t in q] for q in typecounts]
```
