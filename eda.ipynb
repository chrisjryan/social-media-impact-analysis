```python
# what range of dates are found in each tweet set?
datemin_pl = tweets_pl['date_posted'].min().date().isoformat()
datemax_pl = tweets_pl['date_posted'].max().date().isoformat()
datemin_sq = tweets_sq['date_posted'].min().date().isoformat()
datemax_sq = tweets_sq['date_posted'].max().date().isoformat()
print 'Date range, Plague: %s to %s' % (datemin_pl, datemax_pl)  
print 'Date range, Square: %s to %s' % (datemin_sq, datemax_sq)  
```

```python
# How many tweets were posted by each account for each date in the date range?

import matplotlib
from matplotlib import pyplot as plt

tweetdates_sq = pandas.Series([td.date() for td in tweets_sq['date_posted']])

plt.bar(tweetdates_sq.value_counts().index, tweetdates_sq.value_counts().values)
plt.xticks(rotation='vertical')
plt.ylabel('Number of tweets')
plt.title('TheSquareFilm, tweets per day')
fig = plt.gcf()
fig.set_size_inches(9,6)
```

```python
tweetdates_pl = pandas.Series([td.date() for td in tweets_pl['date_posted']])

datemin = min(tweetdates_pl.value_counts().index)
datemax = max(tweetdates_pl.value_counts().index)
countmax = max(tweetdates_pl.value_counts().values)
dt = datetime.timedelta(days=180)

while datemin < datemax:
    plt.bar(tweetdates_pl.value_counts().index, tweetdates_pl.value_counts().values)
    plt.xticks(rotation='vertical')
    plt.ylabel('Number of tweets')
    plt.title('SurviveAPlague, tweets per day')
    fig = plt.gcf()
    fig.set_size_inches(15,4)
    plt.axis((datemin,datemin+dt,0,countmax))
    plt.show()
    datemin += dt
```

```python
# Functions to create boolean vectors for segmenting 
# the tweet data based on tweet content:

def photo_embedded(media_list):
    try:
        for m in media_list:
            if m['type'] == 'photo':
                return True
    except TypeError: 
        return False
    
def get_boolvecs_phu(tweets):
    """
    Segment the data into disjoint sets depending on their features.

    'tweets' is a Pandas DataFrame of tweets loaded from a json file.
    """

    contains_photo = [photo_embedded(ml) for ml in tweets['media']]
    contains_hashtags = [h != [] for h in tweets['hashtags']]
    contains_url = [any(u) for u in tweets['urls']]
    all_boolvec = zip(contains_photo, contains_hashtags, contains_url)

    contains_photo_only    = [p and all((not h,not u)) for p,h,u in all_boolvec]
    contains_hashtags_only = [h and all((not p,not u)) for p,h,u in all_boolvec]
    contains_url_only      = [u and all((not p,not h)) for p,h,u in all_boolvec]

    contains_hashtags_url_only   = [all((h,u)) and not p for p,h,u in all_boolvec]
    contains_photo_url_only      = [all((p,u)) and not h for p,h,u in all_boolvec]
    contains_photo_hashtags_only = [all((p,h)) and not u for p,h,u in all_boolvec]

    contains_all =  [all((p,h,u)) for p,h,u in all_boolvec]
    contains_none = [any(phu) == False for phu in zip(contains_photo, contains_hashtags, contains_url)]
    
    return contains_photo_only, contains_hashtags_only, contains_url_only, \
        contains_hashtags_url_only, contains_photo_url_only, contains_photo_hashtags_only, \
        contains_all, contains_none
```

```python
# What characterizes tweets with:
# (A) 0 retweets (B) between 1 and 5 retweets (C) >5 retweets?

# note: 'pie' plots in pandas.Series.plot() require pandas v0.14 or later.

# get boolean vectors to segment the tweets based on retweet count:
retweets_0 = tweets_sq['n_retweets'] == 0
retweets_1to5 = [all((t>0, t<=5)) for t in tweets_sq['n_retweets']]
retweets_morethan5 = tweets_sq['n_retweets'] > 5

# i.   segment the tweet set based on retweet count
# ii.  takes subsets of each of these based on the tweet content
# iii. count the total in each subset:
typecounts_0retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_0])]
typecounts_1to5retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_1to5])]
typecounts_morethan5retweets = [sum(s) for s in get_boolvecs_phu(tweets_sq[retweets_morethan5])]

# make the pie charts:
idx = ['photo only', 'hashtags only', 'url only', 'hashtags & url', \
       'photo & url', 'photo & hashtags', 'all 3', 'none']
fig, axes = plt.subplots(nrows=3)
fig.set_size_inches(4,15)
axes[0].pie(typecounts_0retweets, labels=idx)
axes[0].set_title('0 retweets')
axes[1].pie(typecounts_1to5retweets, labels=idx)
axes[1].set_title('1 to 5 retweets')
axes[2].pie(typecounts_morethan5retweets, labels=idx)
axes[2].set_title('>5 retweets')
```
